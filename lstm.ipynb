```python
import pandas as pd
import torch
```

```python
df_train_grouped = pd.read_hdf('data3.h5')
df_train_grouped.head()
```

```python
df_train = pd.read_csv('af2_dataset_training_labeled.csv.gz')
df_test = pd.read_csv('af2_dataset_testset_unlabeled.csv.gz')
```

```python
df_train.head()
df_test.head()
```

```python
def reference_embedding(row):
    protein_name = row['entry']
    grouped_row = df_train_grouped.loc[df_train_grouped['entry'] == protein_name]
    protein_embedding = grouped_row['embeddings'].to_numpy()[0]
    index = row['entry_index']
    #print(protein_embedding)
    return protein_embedding[index]
```

```python
df_train_grouped.loc[df_train_grouped['entry'] == 'GEMI5_HUMAN']['embeddings'].to_numpy()[0][0]

df_train['embeddings'] = df_train.apply(reference_embedding, axis =1)
```
